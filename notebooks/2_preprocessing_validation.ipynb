## 2. Pengujian Pembersihan Teks (`clean_text`)

Fungsi ini melakukan:
- Lowercasing
- Penggantian `--` dengan spasi
- Penghapusan tanda baca & karakter khusus
- Penghapusan stopword masakan khas Indonesia (takaran, kata hubung, dll).

```python
sample_texts = [
    'Ayam Woku Manado -- 1 Ekor Ayam Kampung (potong 12) -- Secukupnya garam & lada -- Lalu tumis',
    'Tahu Goreng Gurih -- 2 sdt Ketumbar bubuk -- 300 ml Air hangat -- Secukupnya garam',
    'Sambal Pedas Gila -- 15 cabe rawit merah -- Secukupnya gula dan garam'
]
for text in sample_texts:
    print('====================================')
    print('ASLI  :', text)
    print('BERSIH:', clean_text(text))
```

## 3. Pengujian Ekstraksi Waktu Memasak (`extract_cooking_time`)

Fungsi ini menggunakan ekspresi reguler (Regex) untuk mendeteksi durasi memasak dalam menit atau jam dari langkah-langkah pembuatan.

```python
test_steps = [
    'Rebus ceker ayam selama 45 menit sampai empuk lalu tiriskan.',
    'Panggang adonan kue di oven selama 1 jam dengan api sedang.',
    'Kukus tahu tempe selama 1.5 jam agar bumbu meresap sempurna.',
    'Tumis bumbu halus selama 10 mnt lalu masukkan santan.',
    'Diamkan adonan dalam kulkas selama 2 jam.',
    'Campurkan telur dengan garam lalu kocok lepas (Tanpa informasi waktu)'
]
for step in test_steps:
    extracted = extract_cooking_time(step)
    print(f'Langkah: "{step[:60]}..."')
    print(f'  => Waktu Terdeteksi: {extracted} menit\n')
```

## 4. Pengujian Klasifikasi Tingkat Kepedasan (`classify_spice_level`)

Menganalisis bahan untuk mendeteksi kepedasan berdasarkan jumlah kata kunci pedas (cabai, lada, merica, dll) dan keberadaan cabai rawit.

```python
test_ingredients_spice = [
    '1 ekor ayam kampung, 2 sendok makan garam, 1 ruas kunyit, bawang putih',
    'tahu, tempe, 1/2 sdt merica bubuk, bawang merah, kecap manis',
    '5 ekor udang, 10 cabe rawit merah, 5 cabai merah besar, sambal korek',
    'telur, 2 cabai keriting merah, bawang merah, bawang putih'
]
for ing in test_ingredients_spice:
    level = classify_spice_level(ing)
    print(f'Bahan: {ing[:80]}...')
    print(f'  => Tingkat Kepedasan: {level}\n')
```

## 5. Pengujian Klasifikasi Tipe Diet (`classify_diet_type`)

Mengklasifikasikan resep sebagai Vegetarian jika tidak mengandung bahan hewani (daging, ayam, ikan, udang, cumi, kaldu ayam, dll).

```python
test_cases = [
    ('1 papan tempe potong dadu, 2 balok tahu, kecap manis, garam, air', 'Tempe'),
    ('250 gr fillet dada ayam, tepung bumbu serbaguna, minyak goreng', 'Ayam'),
    ('5 butir telur bebek, daun jeruk, santan encer, kunyit bubuk', 'Telur'),
    ('ikan gurame 1 ekor, jeruk nipis, bawang putih, garam', 'Ikan'),
    ('tahu sutra, jamur tiram, kaldu jamur, wortel, daun bawang', 'Tahu')
]
for ing, cat in test_cases:
    diet = classify_diet_type(ing, cat)
    print(f'Bahan: {ing[:80]}... (Kategori: {cat})')
    print(f'  => Tipe Diet: {diet}\n')
```
